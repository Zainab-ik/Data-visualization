```python
import matplotlib.pyplot as plt
```

```python
import random
```

```python
dataX=[random.randrange(100) for x in range (100)]
dataY=[random.randrange(100) for x in range (100)]
plt.scatter(dataX,dataY,color="black",marker="*") 
plt.title("scatter star") 
plt.xlabel("A")
plt.ylabel("B") 
for x,y in zip(dataX,dataY):
    plt.annotate('o',xy=(x,y),xytext=(5,-5),textcoords='offset pixels')  
```
